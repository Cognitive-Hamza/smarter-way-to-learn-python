## chp11 ELSE and ELIF statements


else:

```python
species = input("name the species: ")
if species == "snake":
  print("Yes, it is venomous!")
else:
  print("It will probably be okay.")
```

it can also written as below:

```python
species = input("name the species: ")
if species == "snake":
  print("Yes, it is venomous!")
if species != "snake":
  print("It will probably be okay.")
```

elif

```python
# Prompt the user to enter their name and store it in a variable.
name = input("Enter your name: ")

# Prompt the user to enter their age and store it as an integer.
age_text = int(input("Enter your age: ")) #text --> int

# Check if the user's age is greater than or equal to 18.
# If yes, display: "Hello [name], you are eligible."
# If no, display: "Sorry [name], you are not eligible."

if age_text >=18:
	print(f'Hello {name}, you are eligible')

elif age_text == 17:
	print(f'Hello {name}, you still have one year')

else:
	print(f'Sorry {name}, you are not eligible')
```

## chp13 IF statements nested

\These consist of main blocks and then sub-blocks per block

```python
# nested if
age = 17
car_ownership_status = "false"
# main if statement
if age >= 18:
# sub-if block
  if(car_ownership_status == "true"):
    print("Drive your car")
# sub-else block
  else:
    print("Work hard, then buy a new car")
# main else statement
else:
  print("You have to grow up first")
```
