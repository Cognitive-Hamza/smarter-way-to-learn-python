## chp16 Lists: Adding and changing elements

Append: this function allows us to add new things of any data type at the end of a list

```python
cities = ["Peshawar", "Multan", "Lahore", "Quetta", "Islamabad", "Karachi"]
```

```python
'''above i have assigned values to the list cities
now i will append new values to that list'''
cities.append("Hyderabad")
print(cities)
```

```python
print(cities[4])
```

```python
'''another way to append is:'''
cities = cities + ["Swat"]
print(cities)
```

```python
'''insert using index'''
cities.insert(0, "Sialkot")
print(cities)
'''this used the index and printed the new city in the index of 0 or made it the first in list'''
```

## chp17 Lists: Taking slices out of them

```python
'''for example:'''
print(cities[2:7])
```

```python
#this is city[2] onward
smaller_list_of_cities = cities[2:]
print(smaller_list_of_cities)
```

```python
#this everything before city[5]
smaller_list_of_cities = cities[:5]
print(smaller_list_of_cities)
```

## chp18 Lists: Deleting and removing elements

```python
# delete
del cities[2]
print(cities)
```

```python
# second method
cities.remove("Karachi")
print(cities)
```

## chp19 Lists: popping elements

```python
# in remove/delete we completely remove elements permanently, pop removes it from the current list but saves it for later when the index is specified
cities = ["Peshawar", "Multan", "Lahore", "Quetta", "Islamabad", "Karachi"]
karachi_index = cities.index("Karachi") # Find the index of 'Karachi'
visited = cities.pop(karachi_index) # Pop the city at that index

print(f"{cities} are to be visited")
print(f"{visited} are visited")
# The 'f' before the opening quotation mark in f"{cities} are to be visited" indicates an f-string (formatted string).
```

## chp20 Tuples

```python
'''Tuples are similar to lists but we cannot be changed which means that we cannot alter them afterwards after they are created.
Parentheses are used rather than square brackets.'''
tup = (1, 2, 9)
print(type(tup), tup)
```

```python
tup = (1)
print(type(tup), tup)
```

as u can see, the data type shown is int, this happens when there is only one value in the tuple, python then considers it as a normal variable. To change this we it is nessacary we add atleast one comma in the tuple

```python
tup = (1,)
tup[0] = 5
# tried to change the tuple so error is shown
print(type(tup), tup)
```

```python
tup = [1]
tup[0] = 4
# now 'tup' is list which can be changed we can identify this by looking at the square brackets
print(type(tup), tup)
```
