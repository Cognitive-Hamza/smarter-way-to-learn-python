##chp24 Changing cases

```python
city_to_check = input("Enter your city: ")
city_to_check = city_to_check.lower()
cleanest_cities = ["cheyenne", "santa fe", "tucson", "great falls", "honolulu"]
for a_clean_city in cleanest_cities:
  if city_to_check == a_clean_city:
    print("It's one of the cleanest cities")
```

Code below is not incorrect less convenient for both coder and user, reason is because the function while taking input from user converts all letters to upper case. Probably nobody will type the whole name in capital letters so we use the '.lower' (code written above) which converts all letter to lower case. This allows ease for user.


```python
city_to_check = input("Enter your city: ")
city_to_check = city_to_check.upper()
cleanest_cities = ["CHEYENNE", "SANTA FE", "TUSCON", "GREAT FALLS", "HONOLULU"]
for a_clean_city in cleanest_cities:
  if city_to_check == a_clean_city:
    print("It's one of the cleanest cities")
```
