##chp26 Dictionaries: How to code one

The structure for a dictionary is similar to the code for creating a list. Both begin with a variable name and equal sign, are in brackets and are seperated by commas.
The major difference is that in a list, each chunk is one thing. In a dictionary, each chunk is a paired key and value.
As shown below.

```python
# The variable name is also the name of the dictionary.
customer_29876 = {"first name": "David", "last name": "Elliott", "address": "4803 Wellesley St."}
print(customer_29876["first name"])
```

##chp27 Dictionaries: How to pick information out of them

```python
# In a dictionary, the code is similar to a list but you pick out an element by specifying its key.
address_of_customer_29876 = customer_29876["address"]
print(address_of_customer_29876)
# the address is now stored in python's memery so whenever we cal the vaiable it will show the address.
```

```python
# Similarly
first_name_of_customer_29876 = customer_29876["first name"]
print(first_name_of_customer_29876)
```

##chp 28 Dictionaries: The versatility of keys and values

```python
#Keys don't have to be strings. They can be numbers too.
rankings = {1: "Omar", 2: "Usman", 3: "Asad", 4: "Ali", 5: "Pervez", 6: "Asim"}
print(rankings[3])
```

```python
#Values can be numbers, too.
rankings1 = {"Omar": 1, "Usman": 2, "Asad": 3, "Ali": 4, "Pervez": 5, "Asim": 6}
print(rankings1["Asim"])
```

```python
#You can mix strings and numbers any way you want.
things_to_remember = {0: "the lowest number", "a dozen": 12, "snake eyes": "a pair of ones", 13: "a baker's dozen"}
'''These are the pairs:
key is the number 0, value is the string "the lowest number"
key is the string "a dozen", value is the number 12
key is the string "snake eyes", value is the string "a pair of ones"
key is the number 13, value is the string "a baker's dozen"
'''
print(things_to_remember["a dozen"])
```

```python
'''When you're defining a dictionary that contains more than two or three key-value pairs,
it's a good idea to break the pairs into separate lines for readability: '''
things_to_remember = {0: "the lowest number",
                      "a dozen": 12,
                      "snake eyes": "a pair of ones",
                      13: "a baker's dozen"
}
print(things_to_remember["snake eyes"])
'''Things to notice:
Each pair is indented.
Though it isn't necessary, I added a comma after the last pair.'''
```
