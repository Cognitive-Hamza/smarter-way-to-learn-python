```python
# printing a series of numbers using for and range
# Example 8:

print("Case 1:")
for i in range(3):  # Print numbers from 0 to 2
    print (i)

print("Case 2:")
for i in range(5, 13):  # Print numbers from 5 to 12
    print (i)

print("Case 3:")
for i in range(5, 20, 4):  # Print numbers from 5 with distace/step 2 and stop before 10
    print (i)
```

```python
# Print Floyd Triangle with user input
# Example 1:

ran = input("How many rows? ")

#
ran = int(ran)
k = 1
for i in range(1, ran+1):
    for j in range(1, i+1):
        print("*", end=" ")
        k = k + 1
    print()
```
