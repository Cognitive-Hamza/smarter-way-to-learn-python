```python
# Example 2: Calculate the average of list of numbers

numbers = [10, 20, 30, 40, 50]

range(10)

[0,1,2,3,4]

# definite iteration
# run loop 5 times because list contains 5 items
sum = 0
for i in numbers:
    sum = sum + i


list_size = len(numbers)
average = sum / list_size
print(average)
```
