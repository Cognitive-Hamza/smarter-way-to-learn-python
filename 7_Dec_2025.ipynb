```python
#Here if we add numbers:
monthly_income = input("Enter your monthly income: ")
yearly_income = monthly_income * 12
print(yearly_income)
```

```python
'''Above rather than the multiplied resut we get 50000 printed 12 time.
To tackle the following issue we assign data types when taking inputs so that such issues are not faced'''
monthly_income = int(input("Enter your monthly income: "))
yearly_income = int(monthly_income * 12)
print(yearly_income)
```
