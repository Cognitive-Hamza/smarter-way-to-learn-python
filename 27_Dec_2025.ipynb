```python
# Q1:
'''. Write a loop that makes seven calls to **`print()`**, so we get on the output the following triangle:
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
# # # # # # # #
'''



for i in range(8):
    print("# # # # # # # #")
```

```python
# Q2: Iterate 0 to 10 using for loop,
for i in range(0, 11):
    print(i)
# do the same using while loop.
i = 0
while i <= 10:
    print(i)
    i += 1

# Q3: Iterate 10 to 0 using for loop,
for i in range(10, -1, -1):
    print(i)
# do the same using while loop.
i = 10
while i >= 0:
    print(i)
    i -= 1
```
