```python
# Print Reverse Floyd Triangle
# Example 12:

for i in range (1,9):            # rows from 1 to 5
    for j in range (8,i-1,-1):   # column range(5,0,-1) = 54321
        print ("*", end=" "),
    print (" ")
```

```python
# regular `for` loop code
# Example 12:

first = [3, 5, 7]
second = [30, 60, 90]
final = []
for i in first:

    for j in second:

        final.append(i*j)
print(final)
```

```python
# regular `for` loop code
# Example 13:

first = [3, 6, 9]
second = [30, 60, 90]
final1 = []
for i in first:
    final1.append(i)
for j in second:
    final1.append(j)
print(final1)
```

The enumerate() function is useful when we wanted to access both value and its index number or any sequence such as list or string.

```python
# Print elements of the list with its index number using the `enumerate()` function
# Example 14:

# In this program, the for loop iterates through the list and displays the
# elements along with its index number.

numbers = [4, 2, 5, 7, 8]
for i ,v in enumerate(numbers):
    print('Numbers[', i, '] =', v)
```

```python
# Printing the elements of the list with its index number using the `range()` function
# Example 15:

numbers = [1, 2, 4, 6, 8, 10, 11, 16]
size = len(numbers)
for i in range(size):
    print('Index:', i, " ", 'Value:', numbers[i])
```

Iterating srings

```python
# For loop with string
# Example 16:

# Method 1:
language = 'Python'
for letter in language:
    print(letter)

# Method 2: using range() function

for i in range(len(language)):
    print(language[i])
```

```python
# Example 17: Access all characters of a string

name = "Alan"
for i in name:
    print(i, end=' ')
```

```python
# Iterate string in reverse order
# Example 18:

name = "Alan"
for i in name[::-1]:
    print(i, end=' ')
```

```python
# Iterate over words in a sentence using the `split()` function.
# Example 19:
dialogue = "Remember, hope is a good thing, maybe the best of things, and no good thing ever dies"
# split on whitespace
for word in dialogue.split():
    print(word)
```
