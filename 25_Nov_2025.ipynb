## chp7 Math Expressions: Eliminating Ambiguity

Python does follow the rules of BODMAS, parentheses are required if you are not specifing what to do first.

```python
# without parentheses
total = 1 + 9 * 3
print(total)
```

```python
# with parentheses
total = (1 + 9) * 3
print(total)
```

```python
total_new = ((2 * 4) + 10) / 3
print(total_new)
```

## chp8 Concatenating text strings

Concatenate: Linking things together in a chain or series. In this we will be breaking down what we are prompting python to print

```python
# using the phrase "Hello, World!"
greeting = "Hello"
adressee = "World"
whole_greeting = greeting + adressee
print(whole_greeting)
```

this above is not the desired output to correct we will break it down further

```python
greeting = "Hello"
separator = ", "
addressee = "World"
punc = "!"
whole_greeting = greeting + separator + addressee + punc
print(whole_greeting)
```

Other possible ways:

```python
print("Hello, " + separator + "World!")
```

or

```python
print(greeting + ", " + addressee + punc)
```

etc.
